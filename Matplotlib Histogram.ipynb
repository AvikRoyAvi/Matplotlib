```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
std_group1_age = np.random.randint(18, 45, (100))
std_group2_age = np.random.randint(15, 40, (100))
```

```python
print(std_group1_age)
print(std_group2_age)
```

```python
plt.hist(std_group1_age)

plt.title('Student group1 age histogram')
plt.xlabel('Age categories')
plt.ylabel('No. of students age')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(std_group1_age, bins, rwidth = 0.8, orientation = 'vertical', color = "r", label = "Student Group1")

plt.title('Student group1 age histogram')
plt.xlabel('Age categories')
plt.ylabel('No. of students age')
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(std_group1_age, bins, rwidth = 0.8, orientation = 'vertical', color = "r", label = "Student Group1")
plt.hist(std_group2_age, bins, rwidth = 0.8, orientation = 'horizontal', color = "y", label = "Student Group2")

plt.title('Student group1 age histogram')
plt.xlabel('Age categories')
plt.ylabel('No. of students age')
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize = (16,9))
bins = [15,20,25,30,35,40,45]
plt.hist([std_group1_age, std_group2_age], bins, rwidth=0.8, histtype = "bar",
         orientation='vertical', color = ["m", "y"], label = ["Student Group1", "Student Group2"])
plt.title('Student group1 age histogram')
plt.xlabel('Age categories')
plt.ylabel('No. of students age')
plt.legend()
plt.show()
```
