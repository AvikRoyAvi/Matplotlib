```python
import matplotlib.pyplot as plt
```

```python
plt.pie([1])
plt.show()
```

```python
colors = 'r'
plt.pie([1], colors = colors)
plt.show()
```

```python
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 
           'Data Sciece']
class1_students = [45, 15, 35, 25, 30]
```

```python
plt.pie(class1_students, labels = classes)
plt.show()
```

```python
explode = [0.03,0,0.1,0,0]           # To slice the perticuler section
colors = ['b', 'm', 'r', 'y', 'g']   # Color of each section
textprops = {"fontsize": 15}         # Font size of text in pie chart

plt.pie(class1_students, labels = classes, explode = explode, colors = colors, autopct = '%0.1f%%',
       shadow = True, radius = 2, startangle = 270, textprops = textprops)
plt.legend()
plt.show()
```
