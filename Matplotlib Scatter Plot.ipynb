```python
x = df_google_play_store_apps['Rating']
y = df_google_play_store_apps['Reviews']
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, y)
plt.title('Google play store apps scatter plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.figure(figsize = (16, 9))
plt.scatter(x, y, c = 'r', marker = '*', s = 100, alpha = 0.5, linewidths = 4, edgecolor = 'y')
plt.title('Google play store apps scatter plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.figure(figsize = (16, 9))
plt.scatter(x, y, c = 'r', marker = '*', s = 100, alpha = 0.5, linewidths = 4, edgecolor = 'y')
plt.scatter(x, df_google_play_store_apps["Installs"], c = "y", marker = "o", s = 100, alpha=0.5, linewidths=4, edgecolors="c" )
plt.title('Google play store apps scatter plot')
plt.xlabel('Rating')
plt.ylabel("Reviews & Installs")
plt.show()
```
